```python
import json
import pandas as pd
import docx
```

```python
with open("C:\\Users\\Luisa Bez\Documents\\Python Scripts\\Travel Advice\\aopa\\aopa\\country_summary.json") as f:
    countries = json.load(f)
```

```python
df = pd.DataFrame(countries)
df.head()
```

```python
# Extract string from Summary list so it doesn't get replaced with NaN after replace_new_line function has been applied
df['Summary_string'] = [','.join(map(str, l)) for l in df['Summary']]
df.drop(columns=["Summary"], inplace = True)
df.head()
```

```python
def remove_whitespace(col):
    col = col.str.strip()
    return col
```

```python
def replace_new_line(col):
    col = col.str.replace("\n","", regex=True)
    return col
```

```python
df = df.apply(replace_new_line)
df = df.apply(remove_whitespace)
df.head()
```

```python
def clean_summary(col):
    col = col.str.replace('<strong>', '')
    col = col.str.replace('</strong>', '')
    col = col.str.replace('<p>', '')
    col = col.str.replace('</p>', '<br><br>')
    col = col.str.replace('<div class="gem-c-govspeak govuk-govspeak direction-ltr" data-module="govspeak">      <div class="call-to-action">', '')
    col = col.str.replace('</a>', '')
    col = col.str.replace('<a href="', '(https://www.gov.uk/')
    col = col.str.replace('">', ') ')
    col = col.str.replace('<li>', '- ')
    col = col.str.replace('</li>', '<br><br>')
    col = col.str.replace('</ul>', '')
    col = col.str.replace('<ul>', '')
    col = col.str.replace('</div>', '')
    return col
```

```python
df = df.apply(clean_summary)
df.head()
```

```python
# Write the df to a word document where the <br> tags can be replaced with actual new lines using ^l

# open an existing document
doc = docx.Document('./Travel_advice2.docx')

# add a table to the end and create a reference variable
# extra row is so we can add the header row
t = doc.add_table(df.shape[0]+1, df.shape[1])

# add the header rows.
for j in range(df.shape[-1]):
    t.cell(0,j).text = df.columns[j]

# add the rest of the data frame
for i in range(df.shape[0]):
    for j in range(df.shape[-1]):
        t.cell(i+1,j).text = str(df.values[i,j])

# save the doc
doc.save('./Travel_advice2.docx')
```
