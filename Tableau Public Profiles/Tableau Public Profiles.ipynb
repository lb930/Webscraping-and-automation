```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
driver = webdriver.Chrome(executable_path=r'C:\Program Files\chromedriver.exe')
wait = WebDriverWait(driver, 10)


#Create lists which can be fed into a DataFrame
name_list = []
views_list = []
favs_list = []

#Get a Tableau Public profile
with driver:
    driver.get("https://public.tableau.com/profile/luisa6565#!/")

    #Wait until the "Load more" button is displayed and scroll to the bottom of the page until it disappears. This ensures that all vizzes have been loaded
    wait.until(EC.presence_of_element_located((By.ID, "load-more-vizzes")))
    while driver.find_element_by_id("load-more-vizzes").is_displayed():
        driver.execute_script("arguments[0].scrollIntoView()", driver.find_element_by_id("load-more-vizzes"))

    #Find the viz container which contains title, views and favourites    
    vizzes = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".viz-container li.media-viz")))
    for viz in vizzes:
        if not viz.is_displayed():
            continue

        #Get the title   
        title = viz.find_element_by_css_selector('[ng-bind="controller.workbook.title"]').text
        name_list.append(title)

        #Get the number of views
        views_count_list = viz.find_elements_by_css_selector('[ng-bind="controller.workbook.viewCount"]')
        views_count = views_count_list[0].text if len(views_count_list) > 0 else 0
        views_list.append(views_count)

        #Get the number of favourites
        number_of_favorites_list = viz.find_elements_by_css_selector('[ng-bind="controller.workbook.numberOfFavorites"]')
        number_of_favorites = number_of_favorites_list[0].text if len(number_of_favorites_list) > 0 else 0
        favs_list.append(number_of_favorites)

        #print(title, views_count, number_of_favorites)
        
    driver.close()

#Create a DataFrame
df = pd.DataFrame({"Title": name_list, "Views": views_list, "Favourites": favs_list})
```

```python
df = df.astype({"Views": "int32", "Favourites": "int32"})
```

```python
df.sort_values(["Favourites", "Views"], ascending = False)
```
