```python
import json
import pandas as pd
```

## Read in data

```python
with open('./ecco4_lang.json') as json_file:
    data = json.load(json_file)
```

```python
df = pd.DataFrame(data)
df.head()
```

## Data prep

The first step drops rows which contain empty lists and therefore no reviews.

```python
df = df[df.astype(str)['Title'] != '[]']
df.head()
```

```python
df["Language"]  = df["Language"].apply(''.join)
```

```python
df = df.set_index(["Product", "Language"])
```

Split each review into a separate row while keeping the associated product and language.

```python
df_review = (df.Review.apply(pd.Series)
              .stack()
              .reset_index(level=2, drop=True)
              .to_frame('Review'))
```
