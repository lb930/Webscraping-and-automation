```python
data = []
geo = []

r = requests.get("https://signaturewebsiteapi.azurewebsites.net/api/locations/GetLocationsGeoJson/all").json()

# Fetch data and geospatial information for all airports
for feature in range(0, len(r["features"])):
    data.append(r["features"][feature]["properties"])
    geo.append(r["features"][feature]["geometry"]["coordinates"])
```

```python
# Store geospatial information inseparate dataset which can be joined on the other df by index position
df_data = pd.DataFrame(data)
df_geo = pd.DataFrame(geo, columns=["Longitude", "Latitude"])
df = df_data.merge(df_geo, left_index=True, right_index=True)
df
```
